```python
import os
import json
import pandas as pd

```

```python
path1=("I:/projects/phonepe/pulse/data/aggregated/insurance/country/india/state/")
aggr_insur_list=os.listdir(path1)

col1={"States":[], "Years":[], "Quarter":[], "Insurance_type":[], "Insurance_count":[], "Insurance_amount":[]}

for state in aggr_insur_list:
    current_states=path1+state+"/"
    aggr_year_list=os.listdir(current_states)

    for year in aggr_year_list:
        current_year=current_states+year+"/"
        aggr_flie_list=os.listdir(current_year)

        for file in aggr_flie_list:
            current_file=current_year+file
            data= open(current_file,"r")
            
            
            A = json.load(data)
            

            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                col1["States"].append(state)
                col1["Years"].append(year)
                col1["Quarter"].append(int(file.strip(".json")))
                col1["Insurance_type"].append(name)
                col1["Insurance_count"].append(count)
                col1["Insurance_amount"].append(amount)

aggregated_insurance=pd.DataFrame(col1)

aggregated_insurance["States"]=aggregated_insurance["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
aggregated_insurance["States"]=aggregated_insurance["States"].str.replace("-"," ")
aggregated_insurance["States"]=aggregated_insurance["States"].str.replace('dadra & nagar haveli & daman & diu','Dadra and Nagar Haveli and Daman and Diu')
aggregated_insurance["States"]=aggregated_insurance["States"].str.title()
            
```

```python
path2=("I:/projects/phonepe/pulse/data/aggregated/transaction/country/india/state/")
aggr_tranc_list=os.listdir(path1)

col2={"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}

for state in aggr_tranc_list:
    current_states= path2+state+"/"
    aggr_year_list=os.listdir(current_states)

    for year in aggr_year_list:
        current_year=current_states+year+"/"
        aggr_flie_list=os.listdir(current_year)
        
        
        for file in aggr_flie_list:
            current_file=current_year+file
            data=open(current_file,'r')

            B=json.load(data)
            for i in B["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                col2["Transaction_type"].append(name)
                col2["Transaction_count"].append(count)
                col2["Transaction_amount"].append(amount)
                col2["States"].append(state)
                col2["Years"].append(year)
                col2["Quarter"].append(int(file.strip(".json"))) 

aggregated_transaction=pd.DataFrame(col2)

aggregated_transaction["States"]=aggregated_transaction["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
aggregated_transaction["States"]=aggregated_transaction["States"].str.replace("-"," ")
aggregated_transaction["States"]=aggregated_transaction["States"].str.replace('dadra & nagar haveli & daman & diu','Dadra and Nagar Haveli and Daman and Diu')
aggregated_transaction["States"]=aggregated_transaction["States"].str.title()

```
